```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.randn(1000)

def MakeHist(data, bins = 10, title = 'Histogram', xlabel = 'x', ylabel ='ylabel'):   
    plt.hist(data, bins)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()

MakeHist(data, bins=10, title='nbins = 10', xlabel='x', ylabel='counts')
MakeHist(data, bins=100, title='nbins = 100', xlabel='x', ylabel='counts')
MakeHist(data, bins=1000, title='nbins = 1000', xlabel='x', ylabel='counts')
```
