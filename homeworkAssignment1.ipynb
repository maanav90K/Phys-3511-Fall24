```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Problem 1, Stirling Approximation

import math

n_val = np.arange(2,20)

fact_ex = np.array([math.log(math.factorial(n)) for n in n_val])
stir_app = np.array([n*np.log(n)-n for n in n_val])

plt.plot(n_val, fact_ex, label='Exact', marker='o')
plt.plot(n_val, stir_app, label='Approximate', linestyle='--')

plt.xlabel('n')
plt.title('Stirling Approximation')

plt.legend()
plt.show()
```

```python
#Problem 5a, 2D matrix array - 1000 entries

matrix = np.random.randn(1000,1000)

print(matrix)
```

```python
#Problem 5b, subplots x and y values

datax = matrix[:, 0]
datay = matrix[:, 1]
fig, axs = plt.subplots(1, 2)

axs[0].hist(datax, bins=30, color='red', alpha=0.4)
axs[0].set_xlabel('X Random Values')

axs[1].hist(datay, bins=30, color='green', alpha=0.4)
axs[1].set_xlabel('Y Random Values')

plt.show
```

```python
#Problem 5c, scatterplot

datax = matrix[:, 0]
datay = matrix[:, 1]

plt.scatter(datax, datay, color='purple', alpha=0.4)
plt.xlabel('Random Values')

plt.show
```

```python
#Problem 5d, 2d histogram

datax = matrix[:, 0]
datay = matrix[:, 1]

plt.hist2d(datax, datay, bins=30, cmap='inferno')
plt.colorbar()
plt.xlabel('x Random Values')
plt.ylabel('y Random Values')

plt.show
```

```python
#Problem 6, cube graph

values = np.linspace(0,1,100)
print(values)
```

```python
new_values = values**3
fig = plt.figure()
axes = fig.add_axes([0.15,0.1,0.8,0.8])
axes.plot(values,new_values)
axes.set_xlabel('x')
axes.set_ylabel('x^3')
axes.set_title('x^3 vs. x');
```
